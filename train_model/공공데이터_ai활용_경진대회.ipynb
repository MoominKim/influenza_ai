```python
weather = pd.read_csv('/workspaces/influenza_ai/OBS_ASOS_DD_20260524123304.csv',encoding='cp949')
weather.head()
```

```python
weather.info()
```

```python
weather=weather.drop(columns=['지점','지점명'])
weather['일시'] = pd.to_datetime(weather['일시'])
```

```python
iso = weather["일시"].dt.isocalendar()

weather["year"] = iso.year
weather["week"] = iso.week
```

```python
weather["평균 풍속(m/s)"] = (
    weather["평균 풍속(m/s)"]
    .interpolate()
)
```

```python
weather_weekly = (
    weather
    .groupby(["year", "week"])
    .agg({
        "평균기온(°C)": "mean",
        "최저기온(°C)": "mean",
        "평균 풍속(m/s)": "mean",
        "평균 상대습도(%)": "mean"
    })
    .reset_index()
)
```

```python
weather_weekly.columns = [
    "year",
    "week",
    "avg_temp",
    "min_temp",
    "avg_wind",
    "avg_humidity"
]
```

```python
weather_weekly["temp_diff"] = (
    weather_weekly["avg_temp"]
    - weather_weekly["min_temp"]
)
```

```python
weather_weekly["week_sin"] = np.sin(
    2 * np.pi * weather_weekly["week"] / 52
)

weather_weekly["week_cos"] = np.cos(
    2 * np.pi * weather_weekly["week"] / 52
)
```

```python
weather_weekly["temp_ma_3"] = (
    weather_weekly["avg_temp"]
    .rolling(window=3, min_periods=1)
    .mean()
)
```

```python
df=df.rename(columns={'미세먼지':'pm10'})
```

```python
df = df.dropna(subset=["ili"])
df['ili'] = df['ili'].astype('float64')
```

```python
df.info()
```

```python
df_base=df.copy()
```

```python
train_df_base = df_base[df_base["year"] < 2024]
test_df_base = df_base[df_base["year"] >= 2024]

X_train_base = train_df_base.drop(columns=["ili", "year"])
y_train_base = train_df_base["ili"]

X_test_base = test_df_base.drop(columns=["ili", "year"])
y_test_base = test_df_base["ili"]
```

```python
%pip install --upgrade pip
%pip install lightgbm
%pip install scikit-learn
```

```python
from lightgbm import LGBMRegressor

model_base_low = LGBMRegressor(
    num_leaves=15,
    n_estimators=300,
    learning_rate=0.03,
    max_depth=5,
    random_state=42
)
model_base_mid = LGBMRegressor(
    num_leaves=31,
    n_estimators=300,
    learning_rate=0.03,
    max_depth=5,
    random_state=42
)
```

```python
model_base_low.fit(X_train_base, y_train_base)
model_base_mid.fit(X_train_base, y_train_base)
```

```python
pred_base_low = model_base_low.predict(X_test_base)
pred_base_mid = model_base_mid.predict(X_test_base)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
)

mae_low = mean_absolute_error(y_test_base, pred_base_low)
rmse_low = mean_squared_error(
    y_test_base,
    pred_base_low
)
r2_low = r2_score(y_test_base, pred_base_low)

print("MAE_low:", mae_low)
print("RMSE_low:", rmse_low)
print("R2_low:", r2_low)

mae_mid = mean_absolute_error(y_test_base, pred_base_mid)
rmse_mid = mean_squared_error(
    y_test_base,
    pred_base_mid
)
r2_mid = r2_score(y_test_base, pred_base_mid)

print("MAE_mid:", mae_mid)
print("RMSE_mid:", rmse_mid)
print("R2_mid:", r2_mid)
```

```python
low_importance = pd.DataFrame({
    "feature": X_train_base.columns,
    "importance": model_base_low.feature_importances_
})
print("low importance",
low_importance.sort_values(
    "importance",
    ascending=False
))
mid_importance = pd.DataFrame({
    "feature": X_train_base.columns,
    "importance": model_base_mid.feature_importances_
})
print("mid importance",
mid_importance.sort_values(
    "importance",
    ascending=False
))
```
